# Getting 2018 MLB Odds Data


We we're going to be doing live data, as in current 2020 MLB Betting Odds and Results, however, due to the COVID-19 Canceling MLB Season, we decided to do 2018 Season. 

We're gathering data from 2018, so instead of Webscraping from ESPN website (Original Plan), we need to convert CSV to something Panda's can handle. We than see that each game was split into VH (Visiting and Home). This means that every 2 entries can be considered 1 Game that includes odds, runs by each inning, final score, and odds of each game.

```python
import pandas as pd
import numpy as np
odds = pd.read_csv("MLBODD2.csv")
pd.set_option("display.max_rows", None, "display.max_columns", None)
visiting_team = odds[::2]
home_team = odds[1::2]
print(visiting_team)
print("-----")
print(home_team)
```

```python
odds = odds.drop(['1st','2nd','3rd','4th','5th','6th','7th','8th','9th','Close','Close OU','Unnamed: 22'], axis = 1)
odds.rename(columns = {'Unnamed: 18':'RL Line', 'Unnamed: 20': 'OU Line'}, inplace = True)
odds
```

```python
# individual team scores
def Team_Data(Team):
    if Team != 'All':
        Home_Scores = odds.loc[odds.Team == Team]
        Home_Index = Home_Scores.index.values
        # finds opponent index
        for i, val in enumerate(Home_Index):
            if val % 2 == 0:
                Home_Index[i] = Home_Index[i] +1
            if val % 2 == 1:
                Home_Index[i] = Home_Index[i] -1
        Op_Scores = odds.iloc[Home_Index.tolist()]
        Team_Games = pd.concat([Home_Scores, Op_Scores]).sort_index()
    elif Team == 'All':
        Team_Games = odds
    Team_Games = Team_Games.reset_index()
    return Team_Games
```

```python
def OU_Bets(What_Team, Value):
    Over_Counter = 0
    Under_Counter = 0
    Even_Counter = 0
    c = 0
    for i, val in enumerate(Team_Data(What_Team)['Open OU']):
        if i % 2 == 0:
            if val == Value:
                c = c + 1
                if Team_Data(What_Team).iloc[i]['Final']+Team_Data(What_Team).iloc[i+1]['Final'] > Value:
                    Over_Counter = Over_Counter + 1
                    #print(Team_Data(What_Team).iloc[i]['Final']+Team_Data(What_Team).iloc[i+1]['Final'])
                elif Team_Data(What_Team).iloc[i]['Final']+Team_Data(What_Team).iloc[i+1]['Final'] < Value:
                    Under_Counter = Under_Counter + 1
                    #print(Team_Data(What_Team).iloc[i]['Final']+Team_Data(What_Team).iloc[i+1]['Final'])
                elif Team_Data(What_Team).iloc[i]['Final']+Team_Data(What_Team).iloc[i+1]['Final'] == Value:
                    Even_Counter = Even_Counter + 1
                    #print(Team_Data(What_Team).iloc[i]['Final']+Team_Data(What_Team).iloc[i+1]['Final'])
    print('Total Occurances :', Over_Counter+Under_Counter+Even_Counter)
    print("----------")
    print('Over :',Over_Counter)
    print('Under :',Under_Counter)
    print('Even :',Even_Counter)
    print("----------")
```

```python
def RL_Bets(What_Team, Fav_or_Und):
    Bet_Win = 0
    Bet_Loss = 0
    Bet_Ties = 0
    if What_Team != 'All':
        indx = np.where(Team_Data(What_Team)['Team'] == What_Team)
        indx1 = np.where(Team_Data(What_Team)['Team'] != What_Team)
        indx2 = np.where(Team_Data(What_Team)['Run Line'] < 0)
        indx3 = np.where(Team_Data(What_Team)['Run Line'] > 0)
        indx4 = np.intersect1d(indx, indx2)
        indx5 = np.intersect1d(indx1, indx3)
        indx6 = np.intersect1d(indx1, indx2)
        indx7 = np.intersect1d(indx, indx3)
        for i, val in enumerate(Team_Data(What_Team)):
            if Fav_or_Und == 'Fav':
                RL_Op_Games = Team_Data(What_Team).iloc[indx5]
                RL_Op_Games = RL_Op_Games.reset_index()
                RL_Games = Team_Data(What_Team).iloc[indx4]
                RL_Games = RL_Games.reset_index()
            elif Fav_or_Und == 'Und':
                RL_Op_Games = Team_Data(What_Team).iloc[indx6]
                RL_Op_Games = RL_Op_Games.reset_index()
                RL_Games = Team_Data(What_Team).iloc[indx7]
                RL_Games = RL_Games.reset_index()
        x = 0
        while (x < len(RL_Games)):
            Diff = RL_Games['Final'].iloc[x] - RL_Op_Games['Final'].iloc[x]
            if Diff > RL_Games['Run Line'].iloc[x]:
                Bet_Win = Bet_Win + 1
            elif Diff < RL_Games['Run Line'].iloc[x]:
                Bet_Loss = Bet_Loss + 1
            elif Diff == RL_Games['Run Line'].iloc[x]:
                Bet_Ties = Bet_Ties + 1
            x = x + 1
        print(What_Team)
        print("Total Games as",Fav_or_Und ," :", Bet_Win + Bet_Loss)
        print("----------")
        print("Bet Hits :" ,Bet_Win)
        print("Bet Losses :" ,Bet_Loss)
        if Bet_Ties > 0:
            print("Bet Ties :" ,Bet_Ties)
        print("----------")
    if What_Team == 'All':
        indx2 = np.where(Team_Data(What_Team)['Run Line'] < 0)
        indx3 = np.where(Team_Data(What_Team)['Run Line'] > 0)
        RL_Fav = Team_Data(What_Team).iloc[indx2]
        RL_Und = Team_Data(What_Team).iloc[indx3]
        RL_Fav = RL_Fav.reset_index()
        RL_Und = RL_Und.reset_index()
        x = 0
        while (x < len(RL_Fav)):
            Diff = RL_Fav['Final'].iloc[x] - RL_Und['Final'].iloc[x]
            if Diff > RL_Fav['Run Line'].iloc[x]:
                Bet_Win = Bet_Win + 1
            elif Diff < RL_Fav['Run Line'].iloc[x]:
                Bet_Loss = Bet_Loss + 1
            elif Diff == RL_Fav['Run Line'].iloc[x]:
                Bet_Ties = Bet_Ties + 1
            x = x + 1   
        print('All Teams Data')
        print("Total Games :", Bet_Win + Bet_Loss)
        print("----------")
        print("Betting Favorite Hits :" ,Bet_Win)
        print("Betting Underdog Hits :" ,Bet_Loss)
        if Bet_Ties > 0:
            print("Betting Ties :" ,Bet_Ties)
        print("----------")
```

```python
def RL_and_OU_Bets(What_Team, Fav_or_Und, O_U):
    Games_OU = Team_Data(What_Team)[Team_Data(What_Team)['Open OU'] == O_U]
    Bet_Win = 0
    Bet_Loss = 0
    if What_Team != 'All':
        indx = np.where(Games_OU['Team'] == What_Team)
        indx1 = np.where(Games_OU['Team'] != What_Team)
        indx2 = np.where(Games_OU['Run Line'] == -1.5)
        indx3 = np.where(Games_OU['Run Line'] == 1.5)
        indx4 = np.intersect1d(indx, indx2)
        indx5 = np.intersect1d(indx1, indx3)
        indx6 = np.intersect1d(indx1, indx2)
        indx7 = np.intersect1d(indx, indx3)
        for i, val in enumerate(Games_OU):
            if Fav_or_Und == 'Fav':
                RL_Op_Games = Games_OU.iloc[indx5]
                RL_Op_Games = RL_Op_Games.reset_index()
                RL_Games = Games_OU.iloc[indx4]
                RL_Games = RL_Games.reset_index()
            elif Fav_or_Und == 'Und':
                RL_Op_Games = Games_OU.iloc[indx6]
                RL_Op_Games = RL_Op_Games.reset_index()
                RL_Games = Games_OU.iloc[indx7]
                RL_Games = RL_Games.reset_index()
        x = 0
        while (x < len(RL_Games)):
            Diff = RL_Games['Final'].iloc[x] - RL_Op_Games['Final'].iloc[x]
            if Diff > RL_Games['Run Line'].iloc[x]:
                Bet_Win = Bet_Win + 1
            elif Diff < RL_Games['Run Line'].iloc[x]:
                Bet_Loss = Bet_Loss + 1
            elif Diff == RL_Games['Run Line'].iloc[x]:
                Bet_Ties = Bet_Ties + 1
            x = x + 1
        print(What_Team)
        print("Total Games as",Fav_or_Und, "and with Over / Under Value", O_U,  " :", Bet_Win + Bet_Loss)
        print("----------")
        print("Bet Hits :" ,Bet_Win)
        print("Bet Losses :" ,Bet_Loss)
        if Bet_Ties > 0:
            print("Bet Ties :" ,Bet_Ties)
        print("----------")
    if What_Team == 'All':
        indx2 = np.where(Games_OU['Run Line'] == -1.5)
        indx3 = np.where(Games_OU['Run Line'] == 1.5)
        RL_Fav = Games_OU.iloc[indx2]
        RL_Und = Games_OU.iloc[indx3]
        RL_Fav = RL_Fav.reset_index()
        RL_Und = RL_Und.reset_index()
        x = 0
        while (x < len(RL_Fav)):
            Diff = RL_Fav['Final'].iloc[x] - RL_Und['Final'].iloc[x]
            if Diff > RL_Fav['Run Line'].iloc[x]:
                Bet_Win = Bet_Win + 1
            elif Diff < RL_Fav['Run Line'].iloc[x]:
                Bet_Loss = Bet_Loss + 1
            elif Diff == RL_Fav['Run Line'].iloc[x]:
                Bet_Ties = Bet_Ties + 1
            x = x + 1
        print('All Teams Data')
        print("Total Games with Over / Under Value", O_U,":", Bet_Win + Bet_Loss)
        print("----------")
        print("Betting Favorite Hits :" ,Bet_Win)
        print("Betting Underdog Hits :" ,Bet_Loss)
        if Bet_Ties > 0:
            print("Betting Ties :" ,Bet_Ties)
        print("----------")
```

```python
What_Team = input("Selected Team's Name (Abreviation or say All for all teams) : ")
Value = float(input("Over / Under Value : "))
if What_Team != 'All':
    Fav_or_Und = input("Fav or Und : ")
if What_Team == 'All':
    Fav_or_Und = ''
print("----------")
print("––– Over / Under Bet –––")
OU_Bets(What_Team, Value)
print("––– Run Line Bets –––")
RL_Bets(What_Team, Fav_or_Und)
print("––– Run Line Bets with Specific Over / Under –––")
RL_and_OU_Bets(What_Team,Fav_or_Und,Value)
#print('All Games')
#print(Team_Data(What_Team))
```
